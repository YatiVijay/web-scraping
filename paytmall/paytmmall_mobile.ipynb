```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.support.ui import Select
from selenium.webdriver import ActionChains
import pandas as pd
import time
```

```python
options = wb.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
site='https://paytmmall.com/shop/search?q=mobiles&from=organic&child_site_id=6&site_id=2&category=6224&page=1'
```

#mobile

```python
links=[]
for j in range(1,29):
  site='https://paytmmall.com/shop/search?q=mobiles&from=organic&child_site_id=6&site_id=2&category=6224&page='+str(j)+''
  wd = wb.Chrome('chromedriver',options=options)
  wd.get(site)
  for i in range(1,9):
    for k in range(1,5):
      try:
        l=wd.find_element_by_xpath('//*[@id="app"]/div/div[3]/div/div/div[3]/div[2]/div['+str(i)+']/div['+str(k)+']/div/a').get_property('href')
        links.append(l)
      except:
        pass
```

```python
print(len(links))
```

```python
prod_name=[]
star=[]
rating=[]
offer_price=[]
MRP=[]
discount=[]
cashback=[]
promocode=[]
cashback_more=[]
product_code=[]
model_id=[]
warranty=[]
seller_name=[]

for i in links:
  wdItem = wb.Chrome('chromedriver',options=options)
  wdItem.get(i)
  try:
    n=wdItem.find_element_by_class_name('NZJI').text
    prod_name.append(n)
  except:
    prod_name.append('Null')
  try:
    op=wdItem.find_element_by_class_name('_1V3w').text
    offer_price.append(op)
  except:
    offer_price.append('Null')
  try:
    mrp=wdItem.find_element_by_class_name('_2b_6').text.split(' ')[1]
    MRP.append(mrp)
  except:
    MRP.append('Null')
  try:
    d=wdItem.find_element_by_class_name('_3llF').text.split('-')[1]
    discount.append(d)
  except:
    discount.append('Null')
  try:
    s=wdItem.find_element_by_class_name('_2dWu').text
    star.append(s)
  except:
    star.append('Null')
  try:
    r=wdItem.find_element_by_class_name('_38Tb').text
    rating.append(r)
  except:
    rating.append('Null')
  try:
    c=wdItem.find_element_by_class_name('_2csj').text.split('â‚¹')[1]
    cashback.append(c)
  except:
    cashback.append('Null')
  try:
    p=wdItem.find_element_by_class_name('PF1t').find_element_by_tag_name('strong').text
    promocode.append(p)
  except:
    promocode.append('Null')
  try:
    cm=wdItem.find_elements_by_class_name('_12vo')[0].text
    cashback_more.append(cm)
  except:
    cashback_more.append('Null')
  try:
    pc=wdItem.find_element_by_class_name('_1CXW').text.split('Product Code\n')[1].split('\n')[0]
    product_code.append(pc)
  except:
    product_code.append('Null')
  try:
    mid=wdItem.find_element_by_class_name('_1CXW').text.split('Model ID\n')[1].split('\n')[0]
    model_id.append(mid)
  except:
    model_id.append('Null')
  try:
    w=wdItem.find_elements_by_class_name('_3gQu')[0].text.split('\n')[1]
    if 'Select' in w:
      w='1 year manufacturer warranty for phone and 6 months for in the box accessories'
  except:
    try:
      w=wdItem.find_elements_by_class_name('nBgT')[0].text.split('\n')[1]
    except:
      w='Null'
  warranty.append(w)
  try:
    sn=wdItem.find_elements_by_xpath('//*[@id="app"]/div/div[3]/div/div[2]/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/div[2]/div[2]')[0].text
  except:
    try:
      sn=wdItem.find_elements_by_xpath('//*[@id="app"]/div/div[3]/div/div[2]/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/div/div')[0].text
      #//*[@id="app"]/div/div[3]/div/div[2]/div/div[2]/div[2]/div[2]/div[3]/div/div[2]/span
    except:
      sn='Null'
  seller_name.append(sn)
```

```python
df={
  'category_link':['https://paytmmall.com/shop/search?q=mobiles&from=organic&child_site_id=6&site_id=2&category=6224']*len(links),
  'prod_link':links,
  'prod_name':prod_name,
  'stars':star,
  'ratings_reviews':rating,
  'offer_price':offer_price,
  'prod_mrp':MRP,
  'discount':discount,
  'currency':['INR']*len(offer_price),
  'cashback':cashback,
  'promocode':promocode,
  'cashback_more':cashback_more,
  'product_code':product_code,
  'model_id':model_id,
  'warranty':warranty,
  'seller_name':seller_name
}
```

```python
dataset=pd.DataFrame(df)
dataset.head(20)
```

```python
import csv
dataset.to_csv('mobile_paytmall.csv')
```
